```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from matplotlib.colors import ListedColormap
def MaxMinNormalization(x,Max,Min):  
    x = (x - Min) / (Max - Min);  
    return x; 

n_samples = 1500
x,y= make_circles(n_samples, factor=0.3, noise=.2);
X=255*MaxMinNormalization(x,np.max(x),np.min(x))

a = np.linspace(0,255,256);b = np.linspace(0,255,256)
Xa,Yb = np.meshgrid(a,b)
points = [point for point in zip(Xa.flat,Yb.flat)]; np.random.shuffle(points)
t=1;P=np.array(points);P=P[0:int(t*65536)]

plt.figure(figsize=(8, 4)) 
plt.subplot(121)
cm_bright = ListedColormap(['Blue', 'Red'])
plt.scatter(X[:,0], X[:,1], s=1, c=y, cmap=cm_bright)

from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(X,y)
pro=model.predict_proba(P);

plt.subplot(122)
cm = plt.cm.get_cmap('RdBu') 
plt.scatter(P[:,0], P[:,1], s=1, c=pro[:,0], cmap=cm)
plt.show()
```
